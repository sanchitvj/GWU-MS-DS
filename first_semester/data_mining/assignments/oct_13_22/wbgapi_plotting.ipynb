```python
import pandas as pd
import seaborn as sns
import numpy as np
import wbgapi as wb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("df_mi.csv")
```

```python
df
```

```python
df = df.dropna()
df
```

```python
df_co2 = df[["economy", "time", "EN.ATM.CO2E.KT", "SP.POP.TOTL"]]
df_co2
```

```python
df_co2["CO2_per_capita"] = df["EN.ATM.CO2E.KT"] / df["SP.POP.TOTL"]
df_co2
```

```python
df_co2 = df_co2.drop(["EN.ATM.CO2E.KT", "SP.POP.TOTL"], axis=1)
df_co2
```

```python
plt.figure(figsize=(15,10))
sns.histplot(df_co2, x=df_co2["CO2_per_capita"], bins=50, kde=True)
```

Above histogram plot is bimodal distribution.

```python
log_df_co2 = np.log(df_co2["CO2_per_capita"])

plt.figure(figsize=(10,8))
sns.histplot(df_co2, x=log_df_co2, kde=True)
```

After log-transformation histogram tends to noraml (gaussian) distribution.

```python
df_gdp = df[["economy", "time", "NY.GDP.MKTP.PP.KD", "SP.POP.TOTL"]]
df_gdp
```

```python
df_gdp["gdp_per_capita"] = df["NY.GDP.MKTP.PP.KD"] / df["SP.POP.TOTL"]
df_gdp
```

```python
df_gdp = df_gdp.drop(["NY.GDP.MKTP.PP.KD", "SP.POP.TOTL"], axis=1)
df_gdp
```

```python
plt.figure(figsize=(30,20))
sns.boxplot(x=df_gdp.economy, y=df_gdp.gdp_per_capita, data=df_gdp)
```

```python
df_gdp_co2 = df_gdp.copy()
df_gdp_co2["CO2_per_capita"] = df_co2["CO2_per_capita"]
df_gdp_co2
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x=df_gdp_co2["gdp_per_capita"], y=df_gdp_co2["CO2_per_capita"], data=df_gdp_co2)
```

With the increasing GDP per capita CO2 emission also increases. Moderate positive correlation.

```python
temp = wb.economy.DataFrame(df_gdp_co2.economy.to_list()) #.metadata.get(id="CAN", )
temp.head()
```

```python
temp = pd.DataFrame(temp, columns=["incomeLevel"])
# temp.reset_index()
# temp.index.values[0]
# temp.index.values.astype
```

```python
df_il = pd.merge(left=df_gdp_co2, right=temp, how="left", left_on="economy", right_on="id")
df_il
```

```python
df_il.incomeLevel.unique()
```

```python
# df_il["income_cat"] = pd.CategoricalIndex(df_il.incomeLevel)
# df_il['incomeLevel'] = df_il['incomeLevel'].astype('category')
# df_il["income_cat"] = pd.Categorical(df_il.incomeLevel)

df_D = df_il.copy()
df_D["income_cat"] = df_D["incomeLevel"].replace(["LMC", "UMC", "HIC"], [1, 2, 3])
# sns.boxplot(x=df_il.economy, y=df_il.incomeLevel, data=df_il)
df_D["income_cat"] = df_D["income_cat"].astype(int)
df_D
```

```python
plt.figure(figsize=(24, 12))
sns.barplot(x=df_D.economy, y=df_D.income_cat, data=df_D)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=df_D.incomeLevel.value_counts(), y=df_D.incomeLevel.value_counts().values, data=df_D)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x=df_D["CO2_per_capita"], data=df_D, hue=df_D.incomeLevel)
```

From above barplot we can infer that low income economies have less CO2 emission per capita, whereas some high income economies have huge CO2 emission per capita.

```python
plt.figure(figsize=(12,7))
sns.scatterplot(y=df_D["CO2_per_capita"], x=df_D["gdp_per_capita"], data=df_D, hue=df_D.incomeLevel)
```

From above scatterplot we can infer about how economies with different income levels have correlation in CO2 emission per capita and GDP per capita. For the low income economies GDP per capita is low and correlated with CO2 emission. Similarly, for high income economies GDP and CO2 emission is showing linear relationship with moderate correlation.
