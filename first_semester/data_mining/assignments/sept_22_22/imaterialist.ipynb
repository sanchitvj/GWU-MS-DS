```python
import pandas as pd
import numpy as np
import json
from pprint import pprint
```

```python
df = pd.read_csv("train.csv")
```

```python
df = df[:5]
df
```

```python
with open("label_descriptions.json") as j:
    data = json.load(j)
```

```python
cat_js = pd.DataFrame.from_dict(data['categories'])
cat_js.head()
```

```python
att_js = pd.DataFrame.from_dict(data['attributes'])
att_js.head()
```

```python
tmp_df = df.copy()
    
for i, row in tmp_df.iterrows():
                   
    tmp_df.at[i, f'name'] = cat_js[cat_js.id == row.ClassId]['name'].item()
    tmp_df.at[i, f'supercategory'] = cat_js[cat_js.id == row.ClassId]['supercategory'].item()
    tmp_df.at[i, f'level'] = cat_js[cat_js.id == row.ClassId]['level'].item()
    
    ls = list(row.AttributesIds.split(",")) #[121, 3, 23123]
    ls = [int(x) for x in ls]

    for j, x in enumerate(ls):
        if x in ls:
            tmp_df.at[i, f'name_{x}'] = att_js[att_js.id == x]['name'].item()
            tmp_df.at[i, f'supercategory_{x}'] = att_js[att_js.id == x]['supercategory'].item()
            tmp_df.at[i, f'level_{x}'] = att_js[att_js.id == x]['level'].item()
        else:
            tmp_df.at[i, f'name_{x}'] = 0
            tmp_df.at[i, f'supercategory_{x}'] = 0
            tmp_df.at[i, f'level_{x}'] = 0
            
tmp_df.head()
```

```python
tmp_df.to_csv("imaterialist.csv")
```
