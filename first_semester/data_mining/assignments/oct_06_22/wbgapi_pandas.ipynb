```python
indexes = ['EN.ATM.CO2E.KT', 'NY.GDP.MKTP.PP.KD', 'SP.POP.TOTL', 'EG.FEC.RNEW.ZS', 'NV.IND.TOTL.ZS', 'SP.URB.TOTL.IN.ZS', 'GC.NFN.TOTL.GD.ZS',
          'NE.EXP.GNFS.ZS', 'GB.XPD.RSDV.GD.ZS', 'TX.MNF.TECH.ZS.UN', 'TX.VAL.TECH.MF.ZS', 'NV.MNF.TECH.ZS.UN', 'NE.TRD.GNFS.ZS']
```

```python
with open('countries.txt', 'r') as f:
    t = f.read()

countries = []
for i in t.split("  "):
    # if len(i) > 5:
    x = i.replace("\n", "").replace('"','')
    if len(x) > 2:
        countries.append(x)

# print(countries)
```

```python
# iso = wb.economy.coder('Albania') #(['Albania', 'Argentina'])

iso_code = []
for i in countries:
    iso_code.append(wb.economy.coder(i))
    
print(iso_code)
```

```python
df = wb.data.DataFrame(indexes, iso_code, time=range(1994, 2022))
```
