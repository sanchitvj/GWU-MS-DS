```python
WBdf=raw_dat.reset_index()
```

```python
#'EN.ATM.CO2E.KT', 
#CO2 emissions (kt) 

value='EN.ATM.CO2E.KT'
CO2=WBdf.query("series == @value").reset_index()
CO2_economy=CO2['economy']
```

```python
# remove non-numerical data and last two coulmn s as they contain only na
CO2=CO2.drop(['index','economy','series','YR2020','YR2021'],  axis=1)
```

```python
CO2_list = CO2.values.tolist()
#print(CO2_list)
grid_points=list(range(1994,2020,1))
```

```python
data_matrix=CO2_list
fd = skfda.FDataGrid(
    data_matrix=data_matrix,
    grid_points=grid_points,
)

fd.plot()
plt.show()
```
