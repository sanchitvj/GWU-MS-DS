```python
# pick one country and show how the moothers work
#CO2_economy
USACO2=logCO2perCap.iloc[CO2_economy[CO2_economy=="USA"].index.tolist()]
```

```python
USARes=pd.DataFrame()
USARes['values']=USACO2.T.values.tolist()
USARes['Years']=grid_points
USARes

data_matrix=CO2_list
X = skfda.FDataGrid(
    data_matrix=USACO2.values.tolist(),
    grid_points=grid_points,
)


fig, axes = plt.subplots(nrows=3, ncols=3)
for n_basis in range(5, 14):
    basis = skfda.representation.basis.BSpline(n_basis=n_basis)
    X_basis = X.to_basis(basis)
    ax = axes.ravel()[n_basis - 5]
    #ax=axes
    fig = X_basis.plot(axes=ax)
    ax.plot(grid_points,USACO2.T.values.tolist(),'o',markersize=2)
    ax.set_title(f"{n_basis} basis functions")

fig.tight_layout()
plt.show()  
```

```python
# calculate the "residual after fitting a curve to
# CO2 in USA 
n_basis=5
basis = skfda.representation.basis.BSpline(n_basis=n_basis)
#evaluate the fucntion at the spesific time points 
X_basis = X.to_basis(basis)
diff=USACO2.T.values.tolist() - X_basis(grid_points)
#diff
x = np.reshape(diff,diff.size) 
mean = numpy.mean(x)
var = numpy.var(x)
ndata = x - mean
acorr = numpy.correlate(ndata, ndata, 'full')[len(ndata)-1:] 
acorr = acorr / var / len(ndata)
print(acorr)
# Adding plot title.
plt.title("Autocorrelation Plot")
# Providing x-axis name.
plt.xlabel("Lags")
# Plotting the Autocorrelation plot.
plt.acorr(x, maxlags = 3)
print("The Autocorrelation plot for the data is:")
plt.grid(True)
plt.show()
```

```python
###### ‘NY.GDP.MKTP.PP.KD’

#Population, total 
value="NY.GDP.MKTP.PP.KD"
GDP=WBdf.query("series == @value").reset_index()
# remove non-numerical data and last two coulmn s as they contain only na
GDP=GDP.drop(['index','economy','series','YR2020','YR2021'],  axis=1)
GDP = np.log(GDP) 
GDP_list = GDP.values.tolist()
#print(CO2_list)
grid_points=list(range(1994,2020,1))
```

```python
#loop over the years rlationship between CO2 and GDP, quadratic 
#loop over the years rlationship between  Data = pd.DataFrame()
# CO2 and GDP, quadratic 
#import statsmodels.formula.api as smf
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
intr=[]
coef=[]
for i in range(0,57):
  #print(i) 
  Data = pd.DataFrame()
  Data['GDP']=GDP.loc[i]
  Data['GDP_squared']=GDP.loc[i]**2
  Data['CO2']=logCO2perCap.loc[i]
  Data=Data.dropna()
  reg = linear_model.LinearRegression()
  reg.fit(Data[['GDP','GDP_squared']], Data['CO2'])
  intr=np.append(intr,reg.intercept_)
  #coef=np.append(coef,reg.coef_)
  coef.append(reg.coef_) 
result_coef = np.stack(coef, axis=0)    
result_intr = np.stack(intr, axis=0)
```

```python
BETA=np.column_stack([result_intr,result_coef])
```

```python
# in what follows I collect the responce and one 
# covariate, GDP into one DataFrame called MainData
#
dummy=logCO2perCap
dummy['economy']=CO2_economy
dummy =dummy.set_index(['economy'])
dummy=dummy.stack()
dummy=dummy.reset_index()
dummy.columns =['economy', 'year', 'CO2']
  
# 57 countries 0..25
#range(0,25,1)
dummy["T"]=np.kron(np.ones(57),range(0,26,1))
MainData=dummy
#adding GDP
dummy=GDP
dummy['economy']=CO2_economy
dummy =dummy.set_index(['economy'])

#missing dataimputation  
dummy = dummy.transpose()
def replace_missing(df):
    #get only numeric columns to dataframe
    df_nu = df.select_dtypes(include=['number'])
    #get only columns with NaNs
    df_nu = df_nu.loc[:, df_nu.isnull().any()]
    return df.fillna(df_nu.median())
dummy=replace_missing(dummy)
 
dummy = dummy.transpose()
dummy=dummy.stack()
dummy=dummy.reset_index()
dummy.columns =['economy', 'year', 'GDP']
```

```python
# MainDataF DataFrame with logtransformed CO2 and 
#logtransformed GDP, along with Year, economy, and T
# a numerical variable for the years starting with 0 

MainDataF=MainData.merge(dummy)
MainDataF.isnull().sum()
```

```python
# OVERALL LINEAR REGRESSION MODEL
dummy=MainDataF
dummy['GDP_squared']=dummy['GDP']**2
OLS = linear_model.LinearRegression()
OLS.fit(dummy[['GDP','GDP_squared']], dummy['CO2'])
```

```python
#here we form a Design matrrix  1 x x^2
# to plot the relationship between GDP and CO2
#NP is the number of grid points

NP=100

x = np.linspace(min(MainDataF['GDP']),
                max(MainDataF['GDP']),num=NP)


X=np.c_[np.ones(NP),x,np.square(x)]
Yplot= np.matmul(BETA,X.transpose())
#plot a few
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

for i in range(0,9):
    ax = axes.ravel()[i]
    #ax.plot(x,Yplot[i])
    Data = pd.DataFrame()
    Data['GDP']=GDP.loc[i]
    Data['GDP_squared']=GDP.loc[i]**2
    Data['CO2']=logCO2perCap.loc[i]
    Data=Data.dropna()
    reg = linear_model.LinearRegression()
    reg.fit(Data[['GDP','GDP_squared']], Data['CO2'])
    y_pred = reg.predict(Data[['GDP','GDP_squared']])
    ax.plot(Data['GDP'],y_pred)
    y2_pred = OLS.predict(Data[['GDP','GDP_squared']])
    ax.plot(Data['GDP'],y2_pred)
    Data.plot.scatter(x='GDP',y='CO2',marker="o",ax=ax)
    ax.set_title(f" for {i}th country ")
 
fig.tight_layout()
plt.show() 
```
